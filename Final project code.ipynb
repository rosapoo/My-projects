Above, we see that there are there are 45 rows for the date we chose, 15-06-2012, one row for each store in the data set. Therefore, we should not make the date the index, as its lack of uniqueness will not create any benefit for our analysis.

Let us do some data type changes for ease of use.

We shall convert the data type of 'Date' from object to datetime. We will also extract the days, months and years from the dates as integers, just in case we need them, and we will create a column with the difference between the first week and each date value so that we can use that time series data. Finally, we will extract the number of days of difference so that we can use that as an integer rather than a timedelta, which will facilitate the linear regression later.

Let us convert 'Store' to an object - even though its values are numbers, they are used for identification and not for counting or arithmetic.

Let us convert 'Holiday_Flag' to a Boolean - even though its values are numbers, they are used for showing whether or not a particular date was a holiday or not. When its data type was an integer, 1 represented a holiday and 0 not a holiday, but after conversion to a boolean, True will represent a holiday and False will represent not a holiday.

```python
walmart['Date_dt'] = pd.to_datetime(walmart['Date'], format = '%d-%m-%Y')

walmart['Store_str'] = walmart['Store'].astype(str)

walmart['Holiday_Flag_bool'] = walmart['Holiday_Flag'].astype(bool)

walmart.dtypes
```

```python
walmart['Date_delta'] = walmart['Date_dt'] - walmart['Date_dt'].min()
walmart
```

```python
walmart['month'], walmart['day'], walmart['year'] = walmart['Date_dt'].dt.month, walmart['Date_dt'].dt.day, walmart['Date_dt'].dt.year
walmart
```

```python
walmart['Date_delta_int'] = walmart['Date_delta'].dt.days
walmart
```

Let us now start to test linear and generalized linear models on individual predictor variables and then multiple variables. With continuous predictor variables, we will use linear regression. When non-continuous predictor variables are involved, we will switch to generalized linear models.

We can use the statsmodel library for this.

For each individual predictor variable, we will create a model for its linear regression on 'Weekly_Sales'. We then use the fit method to fit the data to the model and finally print the summary of the results. In the summary, we will particularly be looking for a high R-squared, which signifies a good fit, and a P value of less than 0.05, which means it is statistically significant.

Looking through the summaries for each predictor variable below, we can see that 'CPI', 'Unemployment', 'Temperature', 'Holiday_Flag_bool' and, for the most part, 'Store_str' have P-values below 0.05, which means that they are statistically significant. Conversely, all of the models have very low R-squared except for 'Store_str'. This signifies that, on their own, most of these predictor variables will not create a good model, but they may work well when combined with others.

```python
import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas.testing as tm
```

```python
model = smf.ols(formula='Weekly_Sales ~ CPI', data=walmart)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula='Weekly_Sales ~ Unemployment', data=walmart)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula='Weekly_Sales ~ Temperature', data=walmart)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula='Weekly_Sales ~ Fuel_Price', data=walmart)
results = model.fit()
print(results.summary())
```

```python
model = smf.ols(formula='Weekly_Sales ~ Date_delta_int',
    data=walmart).fit()
print(model.summary())
```

```python
print(walmart.Holiday_Flag_bool.unique())
# We need to create dummy variables in order to use the binary predictor variable Holiday_Flag_bool,
# one of which will get dropped in the model in order to avoid collinearity.

model = smf.ols(formula='Weekly_Sales ~ Holiday_Flag_bool',
    data=walmart).fit()
print(model.summary())
```

```python
print(walmart.Store_str.unique())
# We need to create dummy variables in order to use the categorical predictor variable Store_str,
# one of which will get dropped in the model in order to avoid collinearity.

model = smf.ols(formula='Weekly_Sales ~ Store_str',
    data=walmart).fit()
print(model.summary())

# This elongates the result summary, but we can see that nearly all of the Store_str dummy variables
# have a P-value of 0.000; since this is less than 0.05, we can ascertain that
# the Store_str variable is indeed statistically significant and should be kept in the final model.
# Furthermore, the R-squared value for this variable is high at 0.917, which indicates that
# this variable is already getting the model close to high accuracy!
```

Now that we have experimented with some models one by one, let us test them side by side to confirm which one is the most suitable.

We start by fitting the eight models we tried, in the same order as before, and assigning a name starting with 'wal' to each.

For each model, we then get an estimate of the intercept and coefficients for each variable.

After that, we plot the results for a visual depiction of how the models estimate parameters in comparison to each other.

Finally, we use the analysis of variance (ANOVA) method to compare the linear models. In ANOVA, we use the sum of squared residuals (SSR) to measure performance, preferring models with lower SSR.

```python
model1='Weekly_Sales ~ CPI + Temperature + Fuel_Price + Unemployment + Date_delta_int + Holiday_Flag_bool + Store_str'
model2='Weekly_Sales ~ Temperature + Fuel_Price + Unemployment + Date_delta_int + Holiday_Flag_bool + Store_str'
model3='Weekly_Sales ~ Temperature + Fuel_Price + Unemployment + Holiday_Flag_bool + Store_str * Date_delta_int'
model4='Weekly_Sales ~ CPI + Temperature + Fuel_Price + Unemployment + Holiday_Flag_bool + Store_str * Date_delta_int'
model5='Weekly_Sales ~ Temperature + Fuel_Price + Unemployment + Holiday_Flag_bool + Date_delta_int * Store_str'
model6='Weekly_Sales ~ Temperature + Fuel_Price + Date_delta_int + Holiday_Flag_bool + Store_str * Unemployment'
model7='Weekly_Sales ~ (Temperature + Fuel_Price + Unemployment + Holiday_Flag_bool + Store_str) * Date_delta_int'
model8='Weekly_Sales ~ (Temperature + Fuel_Price + Unemployment + Holiday_Flag_bool + Date_delta_int) * Store_str'

wal1=smf.ols(model1, data=walmart).fit()
wal2=smf.ols(model2, data=walmart).fit()
wal3=smf.ols(model3, data=walmart).fit()
wal4=smf.ols(model4, data=walmart).fit()
wal5=smf.ols(model5, data=walmart).fit()
wal6=smf.ols(model6, data=walmart).fit()
wal7=smf.ols(model7, data=walmart).fit()
wal8=smf.ols(model8, data=walmart).fit()
```

```python
mod_results = pd.concat([wal1.params, wal2.params, wal3.params,

          wal4.params, wal5.params, wal6.params, wal7.params, wal8.params], axis=1).rename(columns=lambda x: 'wal' \
                            + str(x + 1)).reset_index().rename(columns={'index': 
                            'param'}).melt(id_vars='param', var_name='model', value_name='estimate')


mod_results
```

```python
fig, ax = plt.subplots()

ax = sns.pointplot(x="estimate", y="param", hue="model",
                   data=mod_results,
                   dodge=True, # jitter the points
                   join=False) # don't connect the points

plt.show()
```

```python
model_names = ['wal1', 'wal2', 'wal3', 'wal4', 'wal5', 'wal6', 'wal7', 'wal8']
walmart_anova = statsmodels.stats.anova.anova_lm(
    wal1, wal2, wal3, wal4, wal5, wal6, wal7, wal8)
walmart_anova.index = model_names
walmart_anova
```

Just as when we had tested the models individually, we see here as well that wal8 (model8) has the best performance. This model is 'Weekly_Sales ~ (Temperature + Fuel_Price + Unemployment + Holiday_Flag_bool + Date_delta_int) * Store_str'.

As our last step, we are going to test the data to avoid overfitting and ensure good overall performance of the model. We want to avoid the possibility of having fit our model so well to our existing training dataset that it cannot fit another dataset of the same variables.

We start by using the patsy module to create design matrices.

After that, we split the data into training and testing sets, each of which contains a portion of the total data.

Then, we fit the linear model and normalize the data so that we can compare coefficients while regularizing, if needed.

Finally, we will check the scores of each split part of the data set. These scores are the R-squared values. If the training and testing data sets have similar, and high scores (between 0 and 1), we can conclude that overfitting - and its opposite, underfitting, in which training data gets bad results but testing data gets good results - have been prevented.

If overfitting or underfitting has taken place, we will need to regularize the model.

```python
from patsy import dmatrices

response, predictors = dmatrices(
    'Weekly_Sales ~ (Temperature + Fuel_Price + Unemployment + Holiday_Flag_bool + Date_delta_int) * Store_str',
    data = walmart
)
```

```python
import sklearn

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(predictors, response,
                                                    random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression(normalize=True).fit(X_train, y_train)



model_coefs = pd.DataFrame(list(zip(predictors.design_info.column_names,

                                    lr.coef_[0])),

                          columns=['variable', 'coef_lr'])

print(model_coefs)
```

```python
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

These scores are the R-squared of the training and testing data, respectively. We can see that they are quite similar and the values are fairly high (as R-squared should be between 0 and 1). So, we can conclude that the data has not been overfitted by the chosen model. Therefore, it is not needed to regularize the data in this case by putting constraints on the coefficients and variables.

We again confirm that our model for weekly sales at the 45 Walmart stores from this dataset during the time period captured in the data is 'Weekly_Sales ~ (Temperature + Fuel_Price + Unemployment + Holiday_Flag_bool + Date_delta_int) * Store_str'.
